```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# total number of schools -> count the names of each individual school
numSchools = len(school_data_complete["school_name"].unique())

# total number of students -> count the names?
numStudents = school_data_complete["student_name"].count()
#numStudents

# total budget -> sum up all the budgets
totalBudget = school_data_complete["budget"].sum()
#totalBudget

# avg math score -> it's .mean(), not .avg() :(
avgMath = school_data_complete["math_score"].mean()
#avgMath

# avg reading score
avgRead = school_data_complete["reading_score"].mean()
#avgRead

# pctage students w/ math score >= 70
passMath = school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]
#passMath
passMathPct = passMath / float(numStudents) * 100
#passMathPct

# pctage students w/ reading score >= 70
passRead = school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
#passRead
passReadPct = passRead / float(numStudents) * 100
#passReadPct

# pctage students w/ both math & reading score >= 70 -> use merge?
passBoth = school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)].count()["student_name"]
#passBoth
passBothPct = passBoth / float(numStudents) * 100
#passBothPct

# put all this data in a dataframe
studentDataDF = pd.DataFrame({
    "Total Schools": [numSchools],
    "Total Students": [numStudents],
    "Total Budget": [totalBudget],
    "Average Math Score": [avgMath],
    "% Students Passing Math": [passMathPct],
    "Average Reading Score": [avgRead],
    "% Students Passing Reading": [passReadPct],
    "% Students Passing Overall": [passBothPct]
})
studentDataDF["Total Students"] = studentDataDF["Total Students"].map("{:,}".format)
studentDataDF["Total Budget"] = studentDataDF["Total Budget"].map("${:,.2f}".format)
studentDataDF["Average Math Score"] = studentDataDF["Average Math Score"].map("{:.2f}".format)
studentDataDF["% Students Passing Math"] = studentDataDF["% Students Passing Math"].map("{:.2f}%".format)
studentDataDF["Average Reading Score"] = studentDataDF["Average Reading Score"].map("{:.2f}".format)
studentDataDF["% Students Passing Reading"] = studentDataDF["% Students Passing Reading"].map("{:.2f}%".format)
studentDataDF["% Students Passing Overall"] = studentDataDF["% Students Passing Overall"].map("{:.2f}%".format)
studentDataDF
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
schoolTypes = school_data.set_index(["school_name"])["type"]
numStudents = school_data_complete["school_name"].value_counts()
totalBudget = school_data_complete.groupby(["school_name"]).mean()["budget"]
perStudentBudget = totalBudget / numStudents

avgMath = school_data_complete.groupby(["school_name"]).mean()["math_score"]
avgRead = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

passMath = school_data_complete[(school_data_complete["math_score"] >= 70)]
passMathPct = passMath.groupby(["school_name"]).count()["student_name"] / numStudents * 100
passRead = school_data_complete[(school_data_complete["reading_score"] >= 70)]
passReadPct = passRead.groupby(["school_name"]).count()["student_name"] / numStudents * 100

passBoth = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
passBothPct = passBoth.groupby(["school_name"]).count()["student_name"] / numStudents * 100

summDF = pd.DataFrame({
    "School Type": schoolTypes,
    "Total Students": numStudents,
    "Total School Budget": totalBudget,
    "Per Student Budget": perStudentBudget,
    "Average Math Score": avgMath,
    "Average Reading Score": avgRead,
    "% Passing Math": passMathPct,
    "% Passing Reading": passReadPct,
    "% Overall Passing": passBothPct
})
summDF["Total Students"] = summDF["Total Students"].map("{:,}".format)
summDF["Total School Budget"] = summDF["Total School Budget"].map("${:,.2f}".format)
summDF["Per Student Budget"] = summDF["Per Student Budget"].map("${:,.2f}".format)
summDF["Average Math Score"] = summDF["Average Math Score"].map("{:,.2f}".format)
summDF["Average Reading Score"] = summDF["Average Reading Score"].map("{:,.2f}".format)
summDF["% Passing Math"] = summDF["% Passing Math"].map("{:,.2f}%".format)
summDF["% Passing Reading"] = summDF["% Passing Reading"].map("{:,.2f}%".format)
summDF["% Overall Passing"] = summDF["% Overall Passing"].map("{:,.2f}%".format)

summDF
#summDF.dtypes
```
